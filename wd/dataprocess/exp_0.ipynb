```python
import json
from collections import Counter

path = "/data/userdata/v-lijingyuan/dpo/final_pairs_diff_1.json"

with open(path, "r", encoding="utf-8") as f:
    data = json.load(f)

print("总实验数量:", len(data))

counter = Counter()

for item in data:
    inp = item.get("input", {})
    name = inp.get("comptation_name") or inp.get("competition_name")
    counter[name] += 1

print("\n每个比赛的样本数量：")
for k, v in counter.items():
    print(f"{k}: {v}")
```

```python
import matplotlib.pyplot as plt

TOP = 75  # 可改成 20 / 50
sorted_items = sorted(counter.items(), key=lambda x: x[1], reverse=True)[:TOP]

names = [k for k, _ in sorted_items]
values = [v for _, v in sorted_items]

plt.figure(figsize=(13, 12))

# 使图表更现代
plt.rcParams["font.size"] = 9
plt.rcParams["axes.edgecolor"] = "#444444"
plt.rcParams["axes.linewidth"] = 1.2

# 绘制横向柱状图
bars = plt.barh(range(len(values)), values)  # 0.6~0.75 都可以微调

# 柱子样式（柔和阴影效果）
for b in bars:
    b.set_linewidth(0.8)
    b.set_edgecolor("#333333")

plt.yticks(range(len(values)), names)
plt.xlabel("Number of Samples", fontsize=13)
plt.title(f"Distribution of RD-Agent Run Counts Across Selected Competitions (40 Focused Tasks out of 75 in MLE-Bench)", fontsize=10, weight="bold")

# 网格线（淡灰）
plt.grid(axis="x", linestyle="--", alpha=0.25)

# 从大到小方向显示（好看）
plt.gca().invert_yaxis()

# 数值标签
for i, v in enumerate(values):
    plt.text(v + max(values)*0.01, i, str(v), va="center", fontsize=10)

plt.tight_layout()

# 导出高清版本
plt.savefig("competition_sample_counts.png", dpi=300, bbox_inches="tight")
plt.savefig("competition_sample_counts.pdf", bbox_inches="tight")

plt.show()
```
