```python
import json
from pathlib import Path
import pandas as pd

BASE_DIRS = {
    #"allowed-parrot": Path("/data/userdata/v-lijingyuan/JobAndExp/amlt_project/logs_allowed-parrot"),
    "creative-shrimp": Path("/data/userdata/v-lijingyuan/JobAndExp/amlt_project/logs_creative-shrimp"),
    "social-crawdad": Path("/data/userdata/v-lijingyuan/JobAndExp/amlt_project/logs_social-crawdad"),
}

rows = []

for exp_name, base_dir in BASE_DIRS.items():
    for json_path in base_dir.rglob("results/grading_report.json"):
        try:
            with open(json_path, "r") as f:
                data = json.load(f)

            rows.append({
                "experiment": exp_name,
                "competition_id": data.get("competition_id"),
                "score": data.get("score"),
                "above_median": data.get("above_median"),
                "any_medal": data.get("any_medal"),
                "gold_medal": data.get("gold_medal"),
                "silver_medal": data.get("silver_medal"),
                "bronze_medal": data.get("bronze_medal"),
            })

        except Exception as e:
            print(f"[ERROR] {json_path}: {e}")

df = pd.DataFrame(rows)

print(df.head())
print("\nTotal runs:", len(df))
```

```python
result = (
    df
    .pivot_table(
        index="competition_id",
        columns="experiment",
        values="score",
        aggfunc="mean"   # 如果同一比赛有多个 seed / run
    )
    .reset_index()
)

#print(result)
```
